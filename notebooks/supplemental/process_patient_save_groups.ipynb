This notebook reads in the folder of IMC TIFFs, and organize them into CSV to be processed by `batch_process_feature.py`

```python
import numpy as np
import pandas as pd
import os
```

```python
import re
IMC_FOLDER = '/project/Xie_Lab/zgu/xiao_multiplex/nsclc_tiff_data2'
roi_list = []
input_file_list = []

for imc_tiff in os.listdir(IMC_FOLDER):
    # Match content between 'tma' and '.tiff'
    match = re.search(r'TMA_(.*?)\.tiff', imc_tiff)

    if match:
        content = match.group(1)
        roi_list.append(content)
        input_file_list.append(os.path.join(IMC_FOLDER, imc_tiff))
    else:
        print("Pattern not found.")
    # break

slide_list = ['NSCLC_ALL']*len(input_file_list)

df_nsclc_all = pd.DataFrame({'SLIDE':slide_list, 'ROI':roi_list, 'input file':input_file_list})

# save processed cytof images in different .pkl bc size constraint
# df_nsclc_all['save_group'] = df_nsclc_all.index // 300
df_nsclc_all['save_group'] = 175 # the leftover pt only

# df_nsclc_all.to_csv('nsclc_group175_only.csv', index=False)
df_nsclc_all
```
